```python
import torch
import pandas as pd
import numpy as np
import networkx as nx
```

```python
df_sample = pd.read_csv("adyen-dataset.csv")
```

```python
edges1 = df_sample.groupby(["ip_address","card_number"]).agg(total_transactions=('has_fraudulent_dispute','size'), total_fraud=('has_fraudulent_dispute','sum')).reset_index()
edges1['type_in'] = 'ip'
edges1['type_out'] = 'card'
edges1.columns=["point_in","point_out","total_transactions","total_fraud","type_in","type_out"]
edges1.sort_values(by="total_transactions", ascending=False)
```

```python
edges2 = df_sample.groupby(["ip_address","email_address"]).agg(total_transactions=('has_fraudulent_dispute','size'), total_fraud=('has_fraudulent_dispute','sum')).reset_index()
edges2['type_in'] = 'ip'
edges2['type_out'] = 'email'
edges2.columns=["point_in","point_out","total_transactions","total_fraud","type_in","type_out"]
edges2.sort_values(by="total_fraud", ascending=False)
```

```python
edges3 =df_sample.groupby(["email_address","card_number"]).agg(total_transactions=('has_fraudulent_dispute','size'), total_fraud=('has_fraudulent_dispute','sum')).reset_index()
edges3['type_in'] = 'email'
edges3['type_out'] = 'card'
edges3.columns=["point_in","point_out","total_transactions","total_fraud","type_in","type_out"]
edges3.sort_values(by="total_transactions", ascending=False)
```

```python
df_edges_n = edges3.copy()
df_edges_n["fraud_ratio"] = df_edges_n["total_fraud"] / df_edges_n["total_transactions"]
avg_fraud_n = df_edges_n.groupby("total_transactions").agg(avg_fraud = ("fraud_ratio", "mean"), count_fraud = ("total_transactions", "sum"))
avg_fraud_n
```

```python
df_edges = pd.concat([edges1,edges2,edges3], axis=0)
df_edges = df_edges[df_edges['total_transactions']>0]
df_edges
```

```python
df_edges["fraud_ratio"] = df_edges["total_fraud"] / df_edges["total_transactions"]
avg_fraud = df_edges.groupby("total_transactions").agg(avg_fraud = ("fraud_ratio", "mean"), count_fraud = ("fraud_ratio", "count"))
avg_fraud
```

```python
import networkx as nx

source = 'point_in'
target = 'point_out'

G = nx.from_pandas_edgelist(df_edges, source=source, target = target, edge_attr='total_transactions')
print(f"number of nodes {len(G.nodes)}")
print(f"number of edges {len(G.edges)}")
```

```python
def category(hashed):
    if (len(df_sample[df_sample["email_address"] == hashed]) > 0):
        return "email_address"
    elif (len(df_sample[df_sample["ip_address"] == hashed]) > 0):
        return "ip_address"
    elif (len(df_sample[df_sample["card_number"] == hashed]) > 0):
        return "card_number"
    else:
        return "Nowhere"


hash = "AsGDA0u3ReB45qkhf3rU5A"
feature = category(hash)
print(feature)
```
